```python
import pandas as pd
import glob
import os
```

```python
population_files = glob.glob("population/*.csv")

data = []
for csv in population_files:
    city_population = pd.read_csv(csv)
    city_population['City'] = os.path.basename(csv).replace(".csv", "")
    data.append(city_population)
    
population = pd.concat(data, ignore_index = True)
population
```

```python
population_1960 = population.loc[population["Year"] == 1960]

population_1960['PopulationPercent'] = population_1960['Population'] / population_1960['Population'].sum()

population_1960.sort_values('Population', ascending=False)
```

```python
#complete population data from 1950 to 2035

population.groupby('Year').City.nunique()
```

```python
weather = pd.read_csv("weather/CostOfColdWeather.csv")

weather_1960 = weather.loc[weather['Year'] == 1960]
weather_1960
```

```python
CostOfCold_1960 = pd.merge(population_1960, weather_1960, on='City')

CostOfCold_1960['CostOfCold'] = CostOfCold_1960['PopulationPercent'] * CostOfCold_1960['TenYearMA']

CostOfCold_1960
```

```python
CostOfCold_1960['CostOfCold'].sum()
```

```python
population['TotalPopulation'] = population['Population'].groupby(population['Year']).transform('sum')
population['PopulationPercent'] = population['Population'] / population['TotalPopulation']

population.head()
```

```python
weather.head()
```

```python
CostOfCold = pd.merge(population, weather, how='inner', left_on=['City', 'Year'], right_on=['City', 'Year'])

CostOfCold['CostOfCold'] = CostOfCold['PopulationPercent'] * CostOfCold['TenYearMA']

CostOfCold.head()
```

```python
CostOfCold.groupby(by=['Year'])['CostOfCold'].sum()
```

```python
CostOfCold_IsolatePopulation = pd.merge(population, weather_1960, how='inner', left_on=['City'], right_on=['City'])

CostOfCold_IsolatePopulation['CostOfCold'] = CostOfCold_IsolatePopulation['PopulationPercent'] * CostOfCold_IsolatePopulation['TenYearMA']

CostOfCold_IsolatePopulation.head(10)
```
